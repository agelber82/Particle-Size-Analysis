```python
%matplotlib notebook
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
```

```python
filename = 'U1484A.csv'
df = pd.read_csv(filename,header=1)
df.keys()
```

```python
plt.figure()
plt.plot(df['%Clay'],-df['Top depth CSF-A (m)'],'-o',lw=1)
plt.title('U1484A', fontsize = 20)
plt.xlabel('%Clay Volume', fontsize=16)
plt.ylabel('Depth below Sea Floor [M]', fontsize=16)
```
